```python
!pip install datasets
from datasets import Dataset
from PIL import Image
import torch
import torchvision.transforms as transforms

# Define image transformation to match LLaVA input size
image_transform = transforms.Compose([
    transforms.Resize((336, 336)),  # Resize to match model's expected input size
    transforms.ToTensor(),  # Convert to tensor
])

# Preprocessing function
def preprocess_data(example):
    from PIL import Image

    # Load and process image
    image = Image.open(example["image"]).convert("RGB")
    image_inputs = processor.image_processor(images=image, return_tensors="pt")

    # Ensure text exists and is tokenized correctly
    text = example["text"].strip()
    if not text:
        text = "Describe the image."

    text_inputs = tokenizer(
        f"<image>\n{text}",
        return_tensors="pt",
        padding="max_length",
        truncation=True,
        max_length=4096,
    )

    # 🔹 Ensure labels are correctly shaped
    labels = torch.tensor([example["blue_caps_count"]], dtype=torch.float32)  # Ensure shape is (1,)

    return {
        "pixel_values": image_inputs["pixel_values"].squeeze(0),  # Shape: (3, 336, 336)
        "input_ids": text_inputs["input_ids"].squeeze(0),  # Shape: (512,)
        "attention_mask": text_inputs["attention_mask"].squeeze(0),
        "labels": labels,  # Shape should be (1,)
    }

# Apply preprocessing
dataset = Dataset.from_list(data)
dataset = dataset.map(preprocess_data)

# # Convert to Hugging Face Dataset
# dataset = Dataset.from_list(data)
# dataset = dataset.map(preprocess_data)

# for batch in dataset:
#     print(type(batch["pixel_values"]))  # Should now be a tensor
#     print(batch["pixel_values"].shape)  # Expected: (3, 336, 336)
#     break

dataset.set_format("torch")  # Ensure tensor format

# Verify tensor output
first_example = dataset[0]
print("Type of pixel_values:", type(first_example["pixel_values"]))
print("Shape of pixel_values:", first_example["pixel_values"].shape)
print("Type of input_ids:", type(first_example["input_ids"]))
print("Shape of input_ids:", first_example["input_ids"].shape)
print("Type of labels:", type(first_example["labels"]))
print("Value of labels:", first_example["labels"])


for batch in dataset:
    print("Pixel Values Shape:", batch["pixel_values"].shape)  # Expected: (3, 336, 336)
    print("Input IDs Shape:", batch["input_ids"].shape)  # Expected: (512,)
    print("First 10 Input IDs:", batch["input_ids"][:10])  # Check if they are not empty
    break
```

```python
from transformers import Trainer
from torch.optim import AdamW

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    eval_dataset=None,  # No validation set provided
    tokenizer=processor,
    optimizers=(AdamW(model.parameters(), lr=5e-5), None)
)

# Start fine-tuning
trainer.train()
```
